### DATA CLEANING

```python
car_df = car_df[car_df['year'].str.isnumeric()]
```

```python
car_df['year'] = car_df['year'].astype(int)
```

```python
car_df = car_df[car_df['Price']!='Ask For Price']
```

```python
car_df['Price'] = car_df['Price'].str.replace(',', '').astype(int)
```

```python
car_df['kms_driven'] = car_df['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car_df = car_df[car_df['kms_driven'].str.isnumeric()]
```

```python
car_df['kms_driven'] = car_df['kms_driven'].astype(int)
```

```python
car_df
```

```python
car_df.info()
```

```python
car_df = car_df[~car_df['fuel_type'].isna()]
```

```python
car_df['name'] = car_df['name'].str.split(' ').str.slice(0,4).str.join(' ')
```

```python
car_df = car_df.reset_index(drop=True)
```

```python
car_df.info()
```

```python
car_df.describe()
```

```python
car_df = car_df[car_df['Price']<6e6].reset_index(drop=True)
```

### Model building

```python
# seperating feature and label
X = car_df.drop(columns='Price')
y = car_df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name', 'company', 'fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                       remainder= 'passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
# for finding rendom state number for max r2_score
scores = []
for i in range(1000):
  X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=i)
  lr = LinearRegression()
  pipe = make_pipeline(column_trans, lr)
  pipe.fit(X_train,y_train)
  y_pred = pipe.predict(X_test)
  scores.append(r2_score(y_test, y_pred))
```

```python
scores[np.argmax(scores)]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans, lr)
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
r2_score(y_test,y_pred)
```
